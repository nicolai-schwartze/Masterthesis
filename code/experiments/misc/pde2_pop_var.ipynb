```python
import sys
sys.path.append("../../post_proc")

import post_proc as pp
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "D:/Nicolai/MA_Data/misc/larger_pop/"

pde2_10_6 = pp.loadExpObject(path + "cipde2_mpj_rep_0_10_6_fe.json")
pde2_10_7 = pp.loadExpObject(path + "cipde2_mpj_rep_0_10_7_fe.json")
```
