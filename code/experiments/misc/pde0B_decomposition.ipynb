```python
import sys
sys.path.append("../../post_proc")

import post_proc as pp
import pickle
import matplotlib.pyplot as plt
import numpy as np

sys.path.append("../../kernels")
import KernelGSin as gsk
import KernelGauss as gak

from matplotlib import cm
```

```python
gakernel = gak.KernelGauss()
lD = -2
uD = 2
without_sine = np.array([[1,2,0,0,0,-np.pi/2],[1,1,0,0,0,-np.pi/2],[1,0.1,0,0,0,-np.pi/2]])
kernel_1 =  np.array([[1,2,0,0,0,-np.pi/2]])
kernel_2 =  np.array([[1,1,0,0,0,-np.pi/2]])
kernel_3 =  np.array([[1,0.1,0,0,0,-np.pi/2]])

fig = plt.figure(figsize=(7,6))

#######################################
#              solution               #
#######################################
ax = fig.add_subplot(1,2,2, projection='3d')
x = y = np.arange(lD, uD+0.01, 0.01)
X, Y = np.meshgrid(x, y)
    
zs0 = np.array([gakernel.solution(without_sine, np.array([x,y]))\
                for x,y in zip(np.ravel(X), np.ravel(Y))])
    
Z = zs0.reshape(X.shape)
ax.plot_surface(X, Y, Z, cmap=cm.gnuplot)
    
ax.set_xlabel("X0")
ax.set_ylabel("X1")
ax.set_zlabel("f(X0,X1)")
ax.set_title("Gauss Influence", fontsize=18, x=0.65)

#######################################
#              kernel 1               #
#######################################
ax = fig.add_subplot(3,2,5, projection='3d')
x = y = np.arange(lD, uD+0.01, 0.01)
X, Y = np.meshgrid(x, y)
    
zs0 = np.array([gakernel.solution(kernel_1, np.array([x,y]))\
                for x,y in zip(np.ravel(X), np.ravel(Y))])
    
Z = zs0.reshape(X.shape)
ax.plot_surface(X, Y, Z, cmap=cm.gnuplot)
    
ax.set_xlabel("X0")
ax.set_ylabel("X1")
ax.set_zlabel("f(X0,X1)")
ax.set_title("Gauss 1", y=0.3, x=-0.05, rotation=90)


#######################################
#              kernel 2               #
#######################################
ax = fig.add_subplot(3,2,3, projection='3d')
x = y = np.arange(lD, uD+0.01, 0.01)
X, Y = np.meshgrid(x, y)
    
zs0 = np.array([gakernel.solution(kernel_2, np.array([x,y]))\
                for x,y in zip(np.ravel(X), np.ravel(Y))])
    
Z = zs0.reshape(X.shape)
ax.plot_surface(X, Y, Z, cmap=cm.gnuplot)
    
ax.set_xlabel("X0")
ax.set_ylabel("X1")
ax.set_zlabel("f(X0,X1)")
ax.set_title("Gauss 2", y=0.3, x=-0.05, rotation=90)

#######################################
#              kernel 3               #
#######################################
ax = fig.add_subplot(3,2,1, projection='3d')
x = y = np.arange(lD, uD+0.01, 0.01)
X, Y = np.meshgrid(x, y)
    
zs0 = np.array([gakernel.solution(kernel_3, np.array([x,y]))\
                for x,y in zip(np.ravel(X), np.ravel(Y))])
    
Z = zs0.reshape(X.shape)
ax.plot_surface(X, Y, Z, cmap=cm.gnuplot)
    
ax.set_xlabel("X0")
ax.set_ylabel("X1")
ax.set_zlabel("f(X0,X1)")
ax.set_title("Gauss 3", y=0.3, x=-0.05, rotation=90)

fig.tight_layout()
plt.savefig("pde0B_gauss_influence.pdf", bbox_inches='tight')
plt.show()
```
