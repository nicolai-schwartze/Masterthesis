```python
import scipy as scipy
fig = plt.figure()
ax = plt.subplot(111)

kernelnumber = []
distance = []
for c in cipde2_ex2_10_1:
    distance.append(c["normL2"])
    kernelnumber.append(c["sol_kernel"].shape[0])

distance = [d for _,d in sorted(zip(kernelnumber,distance))]
kernelnumber.sort()
pde2, = ax.plot(kernelnumber, distance, color="#c70000", marker="v", linestyle = 'None')

z = np.polyfit(kernelnumber, distance, 1)
p = np.poly1d(z)
ax.plot(np.arange(3,19.1,0.1),p(np.arange(3,19.1,0.1)), "#c70000")
_, _, pde2r, _, _ = scipy.stats.linregress(kernelnumber, distance)

kernelnumber = []
distance = []
for c in cipde3_ex2_10_1:
    distance.append(c["normL2"])
    kernelnumber.append(c["sol_kernel"].shape[0])

distance = [d for _,d in sorted(zip(kernelnumber,distance))]
kernelnumber.sort()
pde3, = ax.plot(kernelnumber, distance, color="#de9414", marker="o", linestyle = 'None')

z = np.polyfit(kernelnumber, distance, 1)
p = np.poly1d(z)
ax.plot(np.arange(3,19.1,0.1),p(np.arange(3,19.1,0.1)), "#de9414")
_, _, pde3r, _, _ = scipy.stats.linregress(kernelnumber, distance)

kernelnumber = []
distance = []
for c in cipde4_ex2_10_1:
    distance.append(c["normL2"])
    kernelnumber.append(c["sol_kernel"].shape[0])

distance = [d for _,d in sorted(zip(kernelnumber,distance))]
kernelnumber.sort()
pde4, = ax.plot(kernelnumber, distance, color="#8402c9", marker="x", linestyle = 'None')

z = np.polyfit(kernelnumber, distance, 1)
p = np.poly1d(z)
ax.plot(np.arange(3,19.1,0.1),p(np.arange(3,19.1,0.1)), "#8402c9")
_, _, pde4r, _, _ = scipy.stats.linregress(kernelnumber, distance)


kernelnumber = []
distance = []
for c in cipde7_ex2_10_1:
    distance.append(c["normL2"])
    kernelnumber.append(c["sol_kernel"].shape[0])

distance = [d for _,d in sorted(zip(kernelnumber,distance))]
kernelnumber.sort()
pde7, = ax.plot(kernelnumber, distance, color="#000000", marker="*", linestyle = 'None')

z = np.polyfit(kernelnumber, distance, 1)
p = np.poly1d(z)
ax.semilogy(np.arange(3,19.1,0.1),p(np.arange(3,19.1,0.1)), "#000000")
_, _, pde7r, _, _ = scipy.stats.linregress(kernelnumber, distance)


box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

ax.legend([pde2, pde3, pde4, pde7], [r"PDE2 $R^2$=" + "%.5f" % pde2r**2, r"PDE3 $R^2$=" + "%.5f" % pde3r**2, \
                                      r"PDE4 $R^2$=" + "%.5f" % pde4r**2, r"PDE7 $R^2$=" + "%.5f" % pde7r**2],\
                                      loc='center right', bbox_to_anchor=(1.52, 0.5))

plt.xlabel('number of kernels')
plt.ylabel('L2 norm')
plt.grid()
plt.xticks(np.arange(2, 20, step=1))
plt.savefig("pde2347_L2norm_kernelNR.pdf", bbox_inches='tight')
plt.show()
```
